```python
# Import dependencies
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# List all files to be combined
file_compliation = ['data/202107.csv', 'data/202108.csv', 'data/202109.csv', 'data/202110.csv', 'data/202111.csv', 'data/202112.csv',
    'data/202201.csv', 'data/202202.csv', 'data/202203.csv', 'data/202204.csv', 'data/202205.csv', 'data/202206.csv', 'data/202207.csv', 'data/202208.csv', 'data/202209.csv', 'data/202210.csv', 'data/202211.csv', 'data/202212.csv',
        'data/202301.csv', 'data/202302.csv', 'data/202303.csv']

# Create an empty list to store each df
df_list = []

# Loop
for file in file_compliation:
    df = pd.read_csv(file)
    df_list.append(df)

# Concatenate all dataframes into one
# https://www.geeksforgeeks.org/how-to-concatenate-two-or-more-pandas-dataframes/
df = pd.concat(df_list)

df
```

```python
df.dtypes
```

```python
## Formatting

# Correcting datatypes
df['started_at'] = pd.to_datetime(df['started_at'], format = "%Y-%m-%d %H:%M:%S")
df['ended_at'] = pd.to_datetime(df['ended_at'], format = "%Y-%m-%d %H:%M:%S")

# Check 
df.dtypes
```

```python
## Formatting

# Lat/Lon decimals
df['start_lat'] = df['start_lat'].round(decimals=5)
df['start_lng'] = df['start_lng'].round(decimals=5)
df['end_lat'] = df['end_lat'].round(decimals=5)
df['end_lng'] = df['end_lng'].round(decimals=5)

# Check
df
```

```python
# Save df to csv file
df.to_csv('data/citibike.csv', index = False)
```
